```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Define functions that computer parallel and perpendicular power 
def P(theta, gamma):
    v = 3e8 * np.sqrt(1 - (1/(gamma**2))) #Compute velocity in terms of a given gamma
    beta = v / (3e8)
    P_par = ((np.sin(theta))**2) / ((1- beta * np.cos(theta))**6)
    P_per = 1 / ((1- beta * np.cos(theta))**4) * (1 - ( ((np.sin(theta))**2) / (gamma**2 * (1 - beta * np.cos(theta))**2)))
    return P_par, P_per 
```

```python
def plot():
    gammas = [1,2,5,10] #Define list of gammas that will be plotted
    thetas = np.linspace(-np.pi,np.pi,10000) #Define array of thetas that span full plane
    plt.figure(figsize=(15, 6))
    for gamma in gammas: #Iterate over each gamma and plot each resulting graph
        
        P_par = []
        P_per = []
    
        #Compute Power for each value of theta
        for theta in thetas:
            P_par.append(P(theta,gamma)[0])
            P_per.append(P(theta,gamma)[1])
    
        #Plot graphs for parallel and perpendicular cases
        
        plt.subplot(1,2,1, projection='polar') #Graph polar plot of theta and power
        plt.polar(thetas,P_par,label=f'gamma = {gamma}')
        plt.yscale('log')
        plt.ylim(1e-3,1e11)
        plt.title('P Parallel')
        plt.legend(loc='upper left')

        plt.subplot(1,2,2, projection='polar')
        plt.polar(thetas,P_per,label=f'gamma = {gamma}')
        plt.title('P Perpendicular')
        plt.yscale('log')
        plt.ylim(1e-3,1e11)
        plt.legend(loc='upper left')
```

```python
plot()
```
